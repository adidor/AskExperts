```python
import heapq

def sort_k_sorted_array(arr, k):
    heap = []
    # Push the first k elements onto the heap
    for i in range(0, min(k, len(arr))):
        heapq.heappush(heap, arr[i])

    # For each element in the array
    for i in range(k, len(arr)):
        #print(i)
        # Replace the smallest element from the heap with the current array element
        arr[i - k] = heapq.heappushpop(heap, arr[i])

    # Extract remaining elements from the heap
    for i in range(len(arr) - k, len(arr)):
        print(i)
        arr[i] = heapq.heappop(heap)

    return arr

A = [1, 2, 3, 6, 4, 5, 7]
print(sort_k_sorted_array(A, 3))  # Expected: [1, 2, 3, 4, 5, 6, 7]
```
